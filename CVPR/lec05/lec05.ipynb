# 1 Fourier Transform 1-d

$$
X(k)=\sum_{n=0}^{N-1}x(n)\exp\{-i\frac{2\pi kn}{N}\},0\leq k\leq N-1
$$

# 2 Fourier Transform 2-d

$$
F(k,l)=\sum_{m=0}^{M-1}\sum_{n=0}^{N-1}f(m,n)\exp\{-i(\frac{2\pi km}{M}+\frac{2\pi ln}{N})\},0\leq k\leq M-1,0\leq l\leq N-1
$$

$$
F(k,l)=|F(k,l)|\exp\{i\ast \Phi(k,l)\}
$$

# 4 Reconstruct Images

## 4.1 Magnitude and Phase Spectrum

```python
magnitude = np.abs(img_dft)
phase = np.angle(img_dft)
reconstructed_dft = magnitude*np.exp(np.complex(0,1)*phase)
reconstructed_img = image_process.inverse_fourier_transform(reconstructed_dft)


magnitude_img = 20*np.log(magnitude)
phase_img = phase*180/np.pi
plt.figure(figsize=(21,7))
plt.subplot(131)
utils.show_img(utils.normalize(magnitude_img))
plt.title('Magnitude Spectrum')
plt.subplot(132)
utils.show_img(utils.normalize(phase_img))
plt.title('Phase Spectrum')
plt.subplot(133)
utils.show_img(utils.normalize(reconstructed_img))
plt.title('Reconstructed Image')
```

## 4.2 High-Pass Filter and Low-Pass Filter

```python
hpf_1 = image_process.generate_circle_hpf(img_dft, 5)
magnitude_1 = np.abs(img_dft)
magnitude_1 = magnitude_1*hpf_1
reconstructed_dft_1 = magnitude_1*np.exp(np.complex(0,1)*phase)
reconstructed_img_1 = image_process.inverse_fourier_transform(reconstructed_dft_1)

hpf_2 = image_process.generate_rect_hpf(img_dft, 5, 1000)
magnitude_2 = np.abs(img_dft)
magnitude_2 = magnitude_2*hpf_2
reconstructed_dft_2 = magnitude_2*np.exp(np.complex(0,1)*phase)
reconstructed_img_2 = image_process.inverse_fourier_transform(reconstructed_dft_2)

hpf_3 = image_process.generate_rect_hpf(img_dft, 1000, 5)
magnitude_3 = np.abs(img_dft)
magnitude_3 = magnitude_3*hpf_3
reconstructed_dft_3 = magnitude_3*np.exp(np.complex(0,1)*phase)
reconstructed_img_3 = image_process.inverse_fourier_transform(reconstructed_dft_3)


plt.figure(figsize=(21,14))
plt.subplot(231)
utils.show_img(utils.normalize(hpf_1))
plt.title('High-Pass Filter 10*10')
plt.subplot(232)
utils.show_img(utils.normalize(hpf_2))
plt.title('High-Pass Filter 10*max')
plt.subplot(233)
utils.show_img(utils.normalize(hpf_3))
plt.title('High-Pass Filter max*10')
plt.subplot(234)
utils.show_img(utils.normalize(reconstructed_img_1))
plt.subplot(235)
utils.show_img(utils.normalize(reconstructed_img_2))
plt.subplot(236)
utils.show_img(utils.normalize(reconstructed_img_3))
```

```python
lpf_1 = image_process.generate_circle_lpf(img_dft, 5)
magnitude_1 = np.abs(img_dft)
magnitude_1 = magnitude_1*lpf_1
reconstructed_dft_1 = magnitude_1*np.exp(np.complex(0,1)*phase)
reconstructed_img_1 = image_process.inverse_fourier_transform(reconstructed_dft_1)

lpf_2 = image_process.generate_rect_lpf(img_dft, 5, 1000)
magnitude_2 = np.abs(img_dft)
magnitude_2 = magnitude_2*lpf_2
reconstructed_dft_2 = magnitude_2*np.exp(np.complex(0,1)*phase)
reconstructed_img_2 = image_process.inverse_fourier_transform(reconstructed_dft_2)

lpf_3 = image_process.generate_rect_lpf(img_dft, 1000, 5)
magnitude_3 = np.abs(img_dft)
magnitude_3 = magnitude_3*lpf_3
reconstructed_dft_3 = magnitude_3*np.exp(np.complex(0,1)*phase)
reconstructed_img_3 = image_process.inverse_fourier_transform(reconstructed_dft_3)


plt.figure(figsize=(21,14))
plt.subplot(231)
utils.show_img(utils.normalize(lpf_1))
plt.title('High-Pass Filter 10*10')
plt.subplot(232)
utils.show_img(utils.normalize(lpf_2))
plt.title('High-Pass Filter 10*max')
plt.subplot(233)
utils.show_img(utils.normalize(lpf_3))
plt.title('High-Pass Filter max*10')
plt.subplot(234)
utils.show_img(utils.normalize(reconstructed_img_1))
plt.subplot(235)
utils.show_img(utils.normalize(reconstructed_img_2))
plt.subplot(236)
utils.show_img(utils.normalize(reconstructed_img_3))
```
